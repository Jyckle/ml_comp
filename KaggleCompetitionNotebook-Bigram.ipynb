### Import Training data and examine

```python
def prepare_train_data(filename):
    rows=[]
    with open(filename) as file:
        for row in file:
            rows.append(row.strip())
    return rows
```

```python
train = prepare_train_data('train.txt')
```

```python
def prepare_test_data(filename):
    choices=[]
    reg_exp = re.compile('([^ ]*) \{(.*)\|(.*)\}')
    with open(filename) as file:
        for row in file:
            #extract options
            match = reg_exp.search(row)
            if match:
                choice = (match.group(1)+ " " + match.group(2),match.group(1) + " " + match.group(3))
            choices.append(choice)
    return choices
```

```python
choices = prepare_test_data('test.txt')
choices[0]
```

```python
def bigram_probability(option_1,option_2,train):
    count_1 = train.count(option_1)
    count_2 = train.count(option_2)
    if count_1==0: count_1 = 1
    if count_2==0: count_2 = 1
    prob_1 = count_1/(count_1+count_2)
    prob_2 = count_2/(count_1+count_2)
    return prob_1, prob_2
```

```python
start = time.time()
prob = bigram_probability(choices[0][0],choices[0][1],full_train)
end = time.time()
print(end-start, prob)
```

```python
expected_runtime = (end-start)*20000
expected_runtime/60
```

### Try Unigram Model

```python
def bigram_model(train, choices):
    results = [['Id','Expected']]
    for index, choice in enumerate(choices):
        c1,c2 = bigram_probability(choice[0],choice[1],train)
        results.append([index,c1])
    return results
```

```python
results = bigram_model(full_train,choices)
```

```python
results
```

```python
out_file = open('unigram_model.csv','w')
count =0
for line in results:
    if count >= 2:
        index = int(line[0])+1
    else:
        index = line[0]
    output = str(index) + "," + str(line[1]) + "\n"
    out_file.write(output)
    count += 1
out_file.close()
    
```
